```python
import cv2
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
from sklearn.linear_model import LinearRegression
sns.set()
%matplotlib auto
```

```python

from PIL import Image
```

```python
ilIndices = [0, 1]
wBIndices = [0, 1]
vpIndices = [0, 1, 2, 3, 4]
surfacesSuffixes = ['w', 'm', 'c', 'y']

srcFolder = './il1/'
destFolder = './cropped/'
cropH, cropW = 20, 20
```

```python
def getSquareSurfaceNames(ilIndex, wBIndex):
    names = []
    for suff in surfacesSuffixes:
        names.append(f'{ilIndex}-w{wBIndex}-{suff}.jpg')
    return names

def getCroppedNames(ilIndex, wBIndex):
    names = []
    for suff in surfacesSuffixes:
        names.append(f'{destFolder}{ilIndex}-w{wBIndex}-{suff}.jpg')
    return names

def getResizedNames(ill, wBIndex):
    names = []
    src = './il' + str(ill) + '/'
    for vpIndex in vpIndices:
        names.append(f'{src}resized/{vpIndex}-w{wBIndex}.jpg')
    return names

def resizeImages():
    for vpI in vpIndices:
        for wBIndex in wBIndices:
            src = f'{srcFolder}{vpI}-w{wBIndex}.jpg'
            dest = f'{srcFolder}resized/{vpI}-w{wBIndex}.jpg'
            img = Image.open(src)
            orW = img.width
            orH = img.height
            
            if orH > orW:
                newH = 1200
                newW = int(newH * (orW / orH))
            else:
                newW = 1600
                newH = int(newW * (orH / orW))
            
            
            pilImg = img.resize((newW, newH), resample = 0)
            pilImg.save(dest)
        
```

```python
def cropImages():
    for ilIndex in ilIndices:
        for wBIndex in wBIndices:
            sourceImages = getSourceNames(ilIndex, wBIndex)
            destImages = getCroppedNames(ilIndex, wBIndex)
            for i in range(0, len(sourceImages)):
                img = cv2.imread(sourceImages[i])
                cropped = img[0:cropH, 0:cropW]
                cv2.imwrite(destImages[i], cropped)
    
```
